```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium import plugins
from folium.plugins import MarkerCluster
from folium.plugins import HeatMap

import warnings
warnings.filterwarnings('ignore')
```

```python
data_paises_limpo = pd.read_csv('pais_limpo.csv')
```

```python
world_map= folium.Map(tiles="cartodbpositron")
marker_cluster = MarkerCluster().add_to(world_map)
```

```python
for i in range(len(data_paises_limpo)):
        lat = data_paises_limpo.iloc[i]['Latitude']
        long = data_paises_limpo.iloc[i]['Longitude']
        radius=20
        popup_text = """-Pais : {} </br>
        
                    </br>-Alunos : {}"""
        popup_text = popup_text.format(data_paises_limpo.iloc[i]['pais'],
                                   data_paises_limpo.iloc[i]['estudantes']
                                   )
        folium.CircleMarker(location = [lat, long], radius=radius, popup= popup_text, fill =True).add_to(marker_cluster)
```
